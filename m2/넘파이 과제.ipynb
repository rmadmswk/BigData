```python
# [과제] zeros_like, ones_like, full_like 함수 사용 예를 작성하세요.
```

```python
# ones_like
test_matrix = np.arange(30).reshape(5,-1)
print(test_matrix)
np.ones_like(test_matrix, np.float32) 
#원본과 동일한 크기로 만들어짐 
# dtype을 지정해주면 원하는 data type의 배열로 생성해서 반환한다.
```

```python
print(test_matrix)
np.zeros_like(test_matrix, np.int8)
# 1로 초기화가 아닌 0으로 초기화
```

```python
print(test_matrix)
np.full_like(test_matrix,4) #(shape, 배열을 채울 값)
# full 함수는 shape 와 채울값 2가지
```

```python
# [과제] 1 ~ 100까지 정수에서 3의 배수인 것만을 출력(for문과 배열 두가지 방식)
```

```python
array = np.arange(1,101)
# case1 배열
arr = array[array%3==0]
arr1 = arr.reshape(1,11,3)
print(arr1,arr1.shape)
# case2 for문
for arr in array:
    if arr % 3 ==0:
        print(arr)
```

```python
# [과제] 1 ~ 100까지 정수에서 5의 배수이면서 2의 배수인 것만을 출력(for문과 배열 두가지 방식)
arr = np.arange(1,101)
# case 1
arr1 = arr[arr%10==0]
arr2 = arr1.reshape(1,2,-1)
print(arr2, arr2.shape)
# case 2
for arr1 in arr:
    if arr1 % 10 == 0:
        print(arr1)
```

```python
# [과제] 표준정규분포로 구성된 5행 5열 다차원 배열을 생성하고 함수를 이용하여 평균과 표준편차를 구하세요.
data = np.random.randn(1,5,5) # 표준정규분포
print(data)
print(np.mean(data)) # 평균
print(np.std(data)) # 표준편차
```
