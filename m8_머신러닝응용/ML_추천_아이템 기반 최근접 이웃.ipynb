```python
import pandas as pd
import numpy as np
movies = pd.read_csv('dataset/ml-latest-small/movies.csv')
ratings = pd.read_csv('dataset/ml-latest-small/ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
rt = pd.pivot(ratings, index='userId', columns='movieId', values='rating')
rt.head()
```

```python
# movies와 ratings 결합
rating_movies = pd.merge(ratings,movies,on='movieId')
rating_movies.drop(['timestamp'],axis=1,inplace=True)
rating_movies.head()
```

```python
ratings_matrix = rating_movies.pivot_table('rating', index='userId',columns='title')
ratings_matrix.head()
```

```python
# NaN값을 모두 0으로 변환
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
```

영화와 영화들 간 유사도 산출
- cosine_similarity 함수는 행을 기준으로 서로 다른 행을 비교해 유사도를 산출한다.
- 영화를 기준으로 cosine_similarity를 적용하면 현재의 rating_matrix가 행 기준이 되고 열 기준이 사용자가 된다.

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T,ratings_matrix_T)
item_sim
```

```python
item_sim_df = pd.DataFrame(data=item_sim, index = ratings_matrix.columns, columns=ratings_matrix.columns )
item_sim_df.head()
```

```python
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[:10]
```
